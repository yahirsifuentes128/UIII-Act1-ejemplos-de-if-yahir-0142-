ejemplo de if

```python
# Ejemplo 1: Clasificación de números (Operadores Relacionales)
# Este ejemplo utiliza if, elif y else para determinar la naturaleza de un número ingresado.

numero = int(input("Ingresa un número: ")) # 1. Lee la entrada y la convierte a entero

if numero > 0:                             # 2. Compara si el número es mayor que 0
    print("El número es positivo.")        # 3. Se ejecuta si la condición 2 es verdadera
elif numero < 0:                           # 4. Si 2 es falso, comprueba si es menor que 0
    print("El número es negativo.")        # 5. Se ejecuta si la condición 4 es verdadera
else:                                      # 6. Se ejecuta si todas las anteriores son falsas
    print("El número es cero.")            # 7. Resultado cuando el número no es ni >0 ni <0
```

```python
# Aquí se utiliza aritmética para calcular un precio final basado en una condición de cantidad.

cantidad = 12                              # 1. Define una variable con la cantidad comprada
precio_unitario = 100                      # 2. Define el precio base de cada unidad

if cantidad >= 10:                         # 3. Verifica si la cantidad es mayor o igual a 10
    total = (cantidad * precio_unitario) * 0.9 # 4. Aplica 10% de descuento (Aritmética: *, -)
    print(f"Total con descuento: {total}") # 5. Muestra el total calculado
else:                                      # 6. Caso para menos de 10 unidades
    total = cantidad * precio_unitario     # 7. Multiplicación simple sin descuento
    print(f"Total sin descuento: {total}") # 8. Muestra el total sin rebaja
```

```python
# Este ejemplo muestra cómo combinar condiciones usando and y or.

edad = 20                                  # 1. Define la edad del usuario
tiene_entrada = False                      # 2. Define si el usuario posee entrada (Booleano)

# Evalúa si es mayor de edad Y si tiene entrada
if edad >= 18 and tiene_entrada:           # 3. Ambas condiciones deben ser verdaderas (and)
    print("Bienvenido al evento.")         # 4. Acción si cumple ambos requisitos
elif edad >= 18 and not tiene_entrada:     # 5. Si es mayor pero NO (not) tiene entrada
    print("Debe comprar una entrada.")     # 6. Instrucción específica para este caso
else:                                      # 7. Si es menor de edad (independiente de la entrada)
    print("Acceso denegado a menores.")    # 8. Mensaje final de restricción
```

```python
# Ejemplo 4: Verificación de paridad y rango (Operador Módulo)
# Determina si un número es par y si está en un rango específico.

num = 15                                   # 1. Define el número a evaluar
# Comprueba si el resto de dividir por 2 es 0 (Módulo)
if num % 2 == 0:                           # 2. Operador aritmético % y relacional ==
    print("El número es par.")             # 3. Resultado si es divisible por 2
else:                                      # 4. Si el resto no es 0
    print("El número es impar.")           # 5. Resultado si no es divisible por 2

# Verifica si el número está entre 10 y 20 usando lógica
if num >= 10 and num <= 20:                # 6. Uso de and para evaluar un rango
    print("Además, está en el rango de 10 a 20.") # 7. Confirmación del rango
```
